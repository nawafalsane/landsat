```python
import glob
import rasterio
import pickle
from shapely.geometry import box
import numpy as np
import pandas as pd
```

```python
lst = [file.split("_") for file in glob.iglob('./asset/*TIF')]
lst
```

```python
lst = sorted(lst, key=lambda x: x[4])
```

```python
lst = ["_".join(i) for i in lst]
```

```python
lst
```

```python
from collections import defaultdict
raster_dict = defaultdict(list)
for elem in lst:
    raster_dict[elem.split("_")[4]].append(elem)
```

```python
def ndvi(b3, b4):
    nir_tiff = rasterio.open(b4)
    band1 = nir_tiff.read(1)

    red_tiff = rasterio.open(b3)
    band2 = red_tiff.read(1)

    np_nir_as32 = band1.astype(np.float32)
    np_colour_as32 = band2.astype(np.float32)

    np.seterr(divide='ignore', invalid='ignore')

    # Calculate the NDVI formula.
    numerator = np.subtract(np_nir_as32, np_colour_as32)
    denominator = np.add(np_nir_as32, np_colour_as32)
    result = np.divide(numerator, denominator)
    return result
```

```python
ndvi_lst = []
for i in raster_dict.items():
    if len(i[1]) == 2:
        mid_lst = sorted(i[1], key=lambda x: x.split("_")[-1])
        ndvi_lst.append([ndvi(mid_lst[0], mid_lst[1]), i[0][:4]])
        
    elif len(i[1]) > 2:
        mid_lst = sorted(i[1], key=lambda x: x.split("_")[-1])
        ndvi_lst.append([ndvi(mid_lst[1], mid_lst[2]), i[0][:4]])
```

```python
def ndvi(b3, b4):
    nir_tiff = rasterio.open(b4)
    band1 = nir_tiff.read(1)

    red_tiff = rasterio.open(b3)
    band2 = red_tiff.read(1)

    np_nir_as32 = band1.astype(np.float32)
    np_colour_as32 = band2.astype(np.float32)

    np.seterr(divide='ignore', invalid='ignore')

    # Calculate the NDVI formula.
    numerator = np.subtract(np_nir_as32, np_colour_as32)
    denominator = np.add(np_nir_as32, np_colour_as32)
    result = np.divide(numerator, denominator)
    return result
```

```python
with open('parrot.pkl', 'wb') as f:
    pickle.dump(ndvi_lst, f)
```
